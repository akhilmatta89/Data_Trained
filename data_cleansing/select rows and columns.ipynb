# select Row Column Based on Condition

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head(10)
```

```python
#now lets check the values where fare is > 100
df[df['Fare']>100]
```

```python
df[df['Fare']>500]
```

```python
df[df['Sex']!='female']
```
