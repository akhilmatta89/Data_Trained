# Seaborn Visualization Library

it is  a python data visualisation library based on matplotlib ,it provides a high level interface for drawing attractive and informational stattstical graphs

its plotting function operates on dataframes and arrays contatining whole datasets and internally perform the necessary sematic mapping and statistical aggregation to produce informative plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x = ['SUN','MON','TUE','WED','THU','FRI','SAT']
y = [5,6.7,4,6,2,4.9,1.8]
```

```python
ax = sns.stripplot(x,y)
ax.set(xlabel='Days', ylabel='Amount Spend')
plt.title('graph')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

scatter and strip plot does same only difference is xabel and y label we give it in 2 line

```python
irisdf = pd.read_csv('iris.csv')
irisdf.head()
```

```python
sns.set(style='whitegrid')
ax = sns.stripplot(x='class',y='sepal length',data=irisdf)
plt.title('iris graph')
plt.show
```

in strip plot one can be numeric and other is categorical

```python
ax = sns.swarmplot(x='class', y = 'sepal length',data=irisdf)
plt.title('swarm plot')
plt.show()
```
